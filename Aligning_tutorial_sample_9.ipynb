```python
import graspologic

import numpy as np
%matplotlib inline
```

```python
# import an random array of (60,2) as the coordinates of 60 points in X
# import an random array of (20,2) as the coordinates of 20 points in Z
# rotate Z in n degree for Y
import matplotlib.pyplot as plt
np.random.normal(1)
X = np.random.rand(60,2)
np.random.poisson(2)
Y= Z = np.random.rand(20,2)
#print(Z)
plt.subplots(1, 1, figsize=(6, 6))
plt.title("random dots, transforation for Y")
plt.scatter(Z[:,0], Z[:,1], label="Z", alpha=0.6, color='r')

i = 0
n = 3.14
while i <= 19:
    a_Z= Z[i,0]
    b_Z = Z[i,1]
    c = (a_Z**2 + b_Z**2)**0.5
    n_Z = np.arctan(b_Z/a_Z)
    a_Y = c* np.cos(n + n_Z) 
    b_Y = c* np.sin(n + n_Z)
    Y[i] = ([a_Y, b_Y])
    i = i+1
#print(Y)

plt.scatter(X[:,0], X[:,1], label="X", alpha=0.6, color='c')
plt.scatter(Y[:,0], Y[:,1], label="Y", alpha=0.6, color='y')
plt.xlim(-1.50, 1.50)
plt.ylim(-1.50, 1.50)
plt.legend()
plt.grid();
```

```python
# Signflips_X, Y & X@Q
from graspologic.align import SignFlips
X_prime_signflips = SignFlips().fit_transform(X, Y)
plt.subplots(1, 1, figsize=(6, 6))
plt.title("Signflips, n_X=60, n_Y=20")
plt.scatter(X[:,0], X[:,1], label="X", alpha=0.6, color='c')
plt.scatter(Y[:,0], Y[:,1], label="Y", alpha=0.6, color='y')
plt.scatter(X_prime_signflips[:,0], X_prime_signflips[:,1], label="X@Q_signflips", alpha=0.3, color='m')
plt.xlim(-1.50, 1.50)
plt.ylim(-1.50, 1.50)
plt.legend()
plt.grid();
```

```python
#sign flips_X & X@Q
X_prime_signflips = SignFlips().fit_transform(X, Y)
plt.subplots(1, 1, figsize=(6, 6))
plt.title("Signflips, n_X=60, n_Y=20")
plt.scatter(X[:,0], X[:,1], label="X", alpha=0.6, color='c')
plt.scatter(X_prime_signflips[:,0], X_prime_signflips[:,1], label="X@Q_signflips", alpha=0.3, color='m')
plt.xlim(-1.50, 1.50)
plt.ylim(-1.50, 1.50)
plt.legend()
plt.grid();
```

```python
#Seedless
from graspologic.align import SeedlessProcrustes

X_prime_seedless = SeedlessProcrustes().fit_transform(X, Y)
plt.subplots(1, 1, figsize=(6, 6))
plt.title("Seedless, n_X=60, n_Y=20")
plt.scatter(X[:,0], X[:,1], label="X", alpha=0.6, color='c')
plt.scatter(Y[:,0], Y[:,1], label="Y", alpha=0.6, color='y')
plt.scatter(X_prime_seedless[:,0], X_prime_seedless[:,1], label="X@Q_seedless_procruses", alpha=0.3, color='r')
plt.xlim(-1.50, 1.50)
plt.ylim(-1.50, 1.50)
plt.legend()
plt.grid();
```

```python
#orthogonal
from graspologic.align import OrthogonalProcrustes

X_prime_orthogonal = SeedlessProcrustes().fit_transform(X, Y)
plt.subplots(1, 1, figsize=(6, 6))
plt.title("Orthogonal, n_X=60, n_Y=20")
plt.scatter(X[:,0], X[:,1], label="X", alpha=0.6, color='c')
plt.scatter(Y[:,0], Y[:,1], label="Y", alpha=0.6, color='y')
plt.scatter(X_prime_orthogonal[:,0], X_prime_orthogonal[:,1], label="X@Q_orthogonal_procruses", alpha=0.3, color='b')
plt.xlim(-1.50, 1.50)
plt.ylim(-1.50, 1.50)
plt.legend()
plt.grid();
```

```python
# plot all the outcomes together
plt.subplots(1, 1, figsize=(6, 6))
plt.scatter(X[:,0], X[:,1], label="X", alpha=0.6, color='c')
plt.title("Aligning, n_X=60, n_Y=20")
plt.scatter(X_prime_signflips[:,0], X_prime_signflips[:,1], label="X@Q_signflips", alpha=0.3, color='m')
plt.scatter(X_prime_seedless[:,0], X_prime_seedless[:,1], label="X@Q_seedless_procruses", alpha=0.3, color='r')
plt.scatter(X_prime_orthogonal[:,0], X_prime_orthogonal[:,1], label="X@Q_orthogonal_procruses", alpha=0.3, color='b')
plt.xlim(-1.50, 1.50)
plt.ylim(-1.50, 1.50)
plt.legend()
plt.grid();
```
